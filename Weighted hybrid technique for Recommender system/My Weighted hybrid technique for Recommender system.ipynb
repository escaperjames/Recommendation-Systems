```python
import pandas as pd
import numpy as np
```

```python
credits = pd.read_csv("tmdb_5000_credits.csv")
```

```python
movies_df= pd.read_csv("tmdb_5000_movies.csv")
```

```python
credits.head()
```

```python
movies_df.head()
```

```python
print("Credits : ",credits.shape)
print("Movies : ", movies_df.shape)
```

```python
#renaming column movie_id to id so that it matches id column of movies_df ;
#this will help us merging to two dataset in thr next step
credits_column_renamed = credits.rename(index=str, columns={"movie_id":"id"})

#merging two datastsets on column id
movies_df_merge = movies_df.merge(credits_column_renamed, on="id")
movies_df_merge.head()
```

```python
#drop irreleavent columns
movies_cleaned_df = movies_df_merge.drop(columns=['homepage','title_x','status','production_countries'])
movies_cleaned_df.head()
```

```python
#to check if there are any null objects
movies_cleaned_df.info()
```

<b><h3>Using Weighted average  each movie's Average Rating</b>




src:http://trailerpark.weebly.com/imdb-rating.html?source=post_page

```python
# Calculate all the components based on the above formula
v = movies_cleaned_df['vote_count']
R = movies_cleaned_df['vote_average']
C = movies_cleaned_df['vote_average'].mean()
#Pandas dataframe.quantile() function return values at the given quantile over requested axis, a numpy.percentile.
#Note : In each of any set of values of a variate which divide a frequency distribution into equal groups, 
#each containing the same fraction of the total population.
m = movies_cleaned_df['vote_count'].quantile(0.70)
```

```python
#applying the formula
movies_cleaned_df["weighted_average"] = (R*v + C*m)/(v+m)
```

```python
movies_cleaned_df.head()
```

```python
#Sorting movies in descending order of weighted average
movie_sorted_ranking = movies_cleaned_df.sort_values('weighted_average', ascending=False)
movie_sorted_ranking[['original_title','vote_count','vote_average','weighted_average','popularity']].head(20)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

weight_average = movie_sorted_ranking.sort_values('weighted_average',ascending=False)
plt.figure(figsize=(12,6))
axis1=sns.barplot(x=weight_average['weighted_average'].head(20),y=weight_average['original_title'].head(20),data=weight_average)
plt.xlim(4, 10)
plt.title('Best Movies by average votes', weight='bold')
plt.xlabel('Weighted Average Score', Weight='bold')
plt.ylabel('Movie Title', Weight='bold')
plt.savefig('best_movies.png')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

popularity = movie_sorted_ranking.sort_values('popularity',ascending=False)
plt.figure(figsize=(12,6))
axis1=sns.barplot(x=weight_average['popularity'].head(20),y=weight_average['original_title'].head(20),data=weight_average)

plt.title('Best Movies by Votes', weight='bold')
plt.xlabel('Score of Popularity', Weight='bold')
plt.ylabel('Movie Title', Weight='bold')
plt.savefig('best_popular.png')
```

<b><h3>Recommendation based on scaled weighted average and popularity score(Priority is given 50% to both)</b>

```python
from sklearn.preprocessing import MinMaxScaler

scaling=MinMaxScaler()
#making to values of wt._avg and popularity to 0 to 1
movie_scaled_df = scaling.fit_transform(movies_cleaned_df[['weighted_average','popularity']])
movie_normalised_df = pd.DataFrame(movie_scaled_df, columns=['weighted_average','popularity'])
movie_normalised_df.head()
```

```python
movies_cleaned_df[['normalized_weight_average','normalized_popularity']]= movie_normalised_df
```

```python
movies_cleaned_df.head()
```

```python
movies_cleaned_df['score'] = movies_cleaned_df['normalized_weight_average'] * 0.5 + movies_cleaned_df['normalized_popularity'] * 0.5
movies_scored_df = movies_cleaned_df.sort_values(['score'], ascending=False)
movies_scored_df[['original_title', 'normalized_weight_average', 'normalized_popularity', 'score']].head(20)
```

```python
scored_df = movies_cleaned_df.sort_values('score', ascending=False)

plt.figure(figsize=(16,6))

ax = sns.barplot(x=scored_df['score'].head(10), y=scored_df['original_title'].head(10), data=scored_df, palette='deep')

#plt.xlim(3.55, 5.25)
plt.title('Best Rated & Most Popular Blend', weight='bold')
plt.xlabel('Score', weight='bold')
plt.ylabel('Movie Title', weight='bold')

plt.savefig('scored_movies.png')
```
